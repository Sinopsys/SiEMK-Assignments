```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

### Чтение данных из .xls файла
bal, brick, d2, d3, d4, dist, floor, totsp, walk - объясняющие переменные для цены (price) квартиры

```python
data = pd.read_excel('Данные к задаче 5 дз.xlsx')

bal = list(data['bal_2'])
brick = list(data['brick_2'])
d2 = list(data['d2_2'])
d3 = list(data['d3_2'])
d4 = list(data['d4_2'])
dist = list(data['dist_2'])
floor = list(data['floor_2'])
price = list(data['price_2'])
totsp = list(data['totsp_2'])
walk = list(data['walk_2'])
```

```python
len(walk)
```

### Очистка данных от записей с пропусками

```python
cols = [bal, brick, d2, d3, d4, dist, floor, price, totsp, walk]
to_be_removed = []
for col in cols:
    for j in range(0, len(col)):
        if np.isnan(col[j]):
            to_be_removed.append(j)
            
to_be_removed = sorted(list(set(to_be_removed)))
for rem in range(len(to_be_removed) - 1, -1, -1):
    for i in range(0, len(cols)):
        cols[i].pop(to_be_removed[rem])

l = len(walk)
for col in cols:
    if len(col) != l:
        print('BAD')
```

```python
def graphs(reg, y):
    plt.scatter(reg.fittedvalues, reg.resid)
    plt.xlabel('fitted')
    plt.ylabel('residuals')
    plt.show()

    plt.scatter(reg.fittedvalues, y)
    plt.xlabel('fitted')
    plt.ylabel('observed')
    plt.show()

    hist, bins = np.histogram(reg.resid, bins=15)
    width = 0.7 * (bins[1] - bins[0])
    center = (bins[:-1] + bins[1:]) / 2
    plt.bar(center, hist, align='center', width=width)
    plt.show()
```

### Построим линейную модель и графики

```python
reg = sm.OLS(endog=price, exog=sm.add_constant(np.array([bal, brick, d2, d3, d4, dist, floor, totsp, walk]).T)).fit()
graphs(reg, price)
```

### Интерпретируем графики
fitted-residuals график в идеале должен быть равномерно распределен по все области

fitted-observed график в идеале должен быть распределн вдоль прямой y = y_hat

гисторграмма остатков в идеале имеет нормальное распределение

Как видно, линейная регрессия дает плохие результаты

### Предскажем цену квартиры по параметрам с помощью линейной регресии

```python
reg.predict(exog=np.array([1, 1, 1, 0, 0, 0, 10, 1, 40, 1]).reshape(1, -1))
```

### Используем полулогарифмическую зависимость

```python
reg2 = sm.OLS(endog=np.log(price), exog=sm.add_constant(np.array([bal, brick, d2, d3, d4, dist, floor, totsp, walk]).T)).fit()
graphs(reg2, np.log(price))
```

В этом случае графики выглядят хоть и не идеально, но лучше:

fitted-residuals график распределен более равномерно

fitted-observed распределение более близко к прямой y = y_hat

гисторграмма остатков более походит на нормальное рспределение

### Предскажем цену квартиры по параметрам с помощью полулогарифмической регресии

```python
np.exp(reg2.predict(exog=np.array([1, 1, 1, 0, 0, 0, 10, 1, 40, 1]).reshape(1, -1)))
```

### Используем логарифмическую зависимость

```python
reg3 = sm.OLS(endog=np.log(price), exog=sm.add_constant(np.array([bal, brick, d2, d3, d4, np.log(dist), floor, np.log(totsp), walk]).T)).fit()
graphs(reg3, np.log(price))
```

В этом случае графики выглядят хоть и не идеально, но лучше:

fitted-residuals график распределен более равномерно

fitted-observed распределение более близко к прямой y = y_hat

гисторграмма остатков более походит на нормальное рспределение

### Предскажем цену квартиры по параметрам с помощью логарифмической модели

```python
np.exp(reg3.predict(exog=np.array([1, 1, 1, 0, 0, 0, np.log(10), 1, np.log(40), 1]).reshape(1, -1)))
```

Примем цену 46.09 как более достоверную среди трех моделей, так как визуальная диагностика показала, что логарифмическая регресия является наиболее подходящей

### Интерпретация коэффициентов логарифмической модели

```python
print(reg3.summary())
```

Увеличение регрессора на 1% приблизительно соответствует увеличению y на C% при прочих равных условиях. Это означает, что, например, коэффициент при x9 ~ 0.1, это значит, что квартиры, от которых можно быстро дойти до метро стоят на 10% дороже. Коэффициент при x1 ~ 0.06, это значит, что квартиры с балконом стоят на 6% дороже. И так далее, для каждого регрессора, который принимает 0 или 1.

Регрессоры dist и totsp (x6 и x8 соответственно) - расстояние до центра города и площадь квартиры.
Знаки при этих регрессорах отличаются, то есть чем расстояние до центра больше, тем цена ниже (обратная зависимость). Чем плозадь больше, тем и цена квартиры больше (что логично).

Видно, что наиболее значимым регрессором является именно площадь квартиры, затем идет расстояние до центра города.
Все остальные регрессоры имеют примерно одинаковую значимость (все - менее 5%, то есть все - значимы).
